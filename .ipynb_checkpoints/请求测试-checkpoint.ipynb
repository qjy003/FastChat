```python
def chat_robot(json_data):
    try:
        logger.info("请求===========================>")

        url = "http://127.0.0.1:18888/question"

        response = requests.post(url, json=json_data)
        
        response_data=response.json()
        print(response_data)
        res_return = json.dumps(response_data,ensure_ascii=False)
        logger.info("返回===========================>")
        logger.info(res_return)

        return res_return

    except Exception as e:
        logger.info(e)
        return ''
```

```python
data=chat_robot(json_data)
```
